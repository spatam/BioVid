```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import autocorrelation_plot

# === CONFIGURAZIONE ===
dataset_folder = "/Users/geofargy95/Downloads/BioVid-main/dataset"  # Modifica con il percorso del dataset
output_folder = "analysis_results"
csv_path = "/Users/geofargy95/Downloads/BioVid-main/allowed_words_cropped.csv"  # Percorso al file CSV

os.makedirs(output_folder, exist_ok=True)

# === CARICAMENTO E ANALISI DEL DATASET ===
data = pd.read_csv(csv_path)
print("Dataset preview:\n", data.head())
print("Available columns:", data.columns)

# Aggiungi lunghezza delle parole se necessario
if 'Word_Syllables' not in data.columns:
    data['Word_Length'] = data['word'].apply(len)

# Statistiche Descrittive
stats = data.describe(include='all').transpose()
stats.to_csv(os.path.join(output_folder, "data_statistics.csv"))
print("Complete statistics saved.")

# === VISUALIZZAZIONI DETTAGLIATE ===

# 1. Distribuzione della lunghezza delle parole
plt.figure(figsize=(10, 6))
sns.histplot(data['Word_Length'], kde=True, bins=10, color='blue')
plt.title("Distribution of Word Length in Characters", fontsize=14)
plt.xlabel("Word Length (characters)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "word_length_distribution.png"))
plt.show()

# 2. Boxplot della lunghezza delle parole
plt.figure(figsize=(8, 6))
sns.boxplot(data['Word_Length'], color='green')
plt.title("Boxplot of Word Length", fontsize=14)
plt.xlabel("Word Length (characters)", fontsize=12)
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "word_length_boxplot.png"))
plt.show()

# 3. Scatter Plot tra Noise_Level e Word_Length (se disponibili)
if 'Noise_Level' in data.columns and 'Word_Length' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=data['Noise_Level'], y=data['Word_Length'], alpha=0.7, color='purple')
    plt.title("Scatter Plot: Noise Level vs Word Length", fontsize=14)
    plt.xlabel("Noise Level (dB)", fontsize=12)
    plt.ylabel("Word Length (characters)", fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, "scatter_noise_wordlength.png"))
    plt.show()

# 4. Heatmap della Matrice di Correlazione
numeric_cols = data.select_dtypes(include=['float', 'int'])
if numeric_cols.shape[1] > 1:
    plt.figure(figsize=(12, 8))
    sns.heatmap(numeric_cols.corr(), annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
    plt.title("Correlation Matrix of Numerical Features", fontsize=14)
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, "correlation_matrix.png"))
    plt.show()

# 5. Autocorrelazione della Lunghezza delle Parole (esempio)
if 'Word_Length' in data.columns:
    plt.figure(figsize=(10, 6))
    autocorrelation_plot(data['Word_Length'])
    plt.title("Autocorrelation of Word Length", fontsize=14)
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, "autocorrelation_word_length.png"))
    plt.show()

# 6. KDE Plot della Lunghezza delle Parole
plt.figure(figsize=(10, 6))
sns.kdeplot(data['Word_Length'], shade=True, color='red')
plt.title("KDE Plot of Word Length", fontsize=14)
plt.xlabel("Word Length (characters)", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "kde_word_length.png"))
plt.show()

# 7. Pairplot per Esplorazione Multivariata
if numeric_cols.shape[1] > 1:
    sns.pairplot(data=numeric_cols, diag_kind='kde')
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, "pairplot_numeric_features.png"))
    plt.show()

# === REPORT COMPLETO ===
with open(os.path.join(output_folder, "report.txt"), "w") as report:
    report.write("BioVid Dataset Complete Analysis Report\n")
    report.write("=======================================\n")
    report.write(f"Total Entries: {len(data)}\n")
    report.write(f"Numeric Features: {list(numeric_cols.columns)}\n")
    if 'User_ID' in data.columns:
        report.write(f"Participants: {data['User_ID'].nunique()}\n")
    report.write("\nComplete Statistics:\n")
    report.write(stats.to_string())
    report.write("\n\nVisualizations saved in the output folder.\n")

print(f"Comprehensive analysis complete. Results saved in {output_folder}")
```

```python
import os
import librosa
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# === CONFIGURAZIONE ===
dataset_folder = "/Users/geofargy95/Downloads/BioVid-main/dataset"  # Percorso della cartella "dataset"
output_folder = "mfcc_analysis_results"
os.makedirs(output_folder, exist_ok=True)

audio_features = []

# Estensioni supportate per i file video
supported_extensions = (".mp4", ".MP4")  # Aggiungi altre estensioni se necessario

# === ESTRAZIONE DELLE CARATTERISTICHE AUDIO ===
for root, dirs, files in os.walk(dataset_folder):
    for file in files:
        if file.endswith(supported_extensions):  # Controlla se il file ha un'estensione supportata
            file_path = os.path.join(root, file)
            try:
                # Carica il file audio
                y, sr = librosa.load(file_path, sr=None)
                
                # Estrai i primi 13 MFCC
                mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13).mean(axis=1)
                
                # Salva le caratteristiche rilevanti
                audio_features.append({
                    "File": file,
                    "Folder": os.path.basename(root),
                    "MFCC1": mfccs[0],
                    "MFCC2": mfccs[1],
                    "MFCC10": mfccs[9],  # Il 10 MFCC
                })
            except Exception as e:
                print(f"Error processing {file_path}: {e}")
                
print(f"Total number of videos analyzed: {len(audio_features)}")

# Crea un DataFrame per analizzare i dati
df_mfcc = pd.DataFrame(audio_features)
df_mfcc.to_csv(os.path.join(output_folder, "mfcc_features.csv"), index=False)
print("MFCC features saved to mfcc_features.csv")

# === GRAFICO COMBINATO DEI 3 MFCC ===
plt.figure(figsize=(14, 6))
sns.lineplot(data=df_mfcc, x=range(len(df_mfcc)), y="MFCC1", label="MFCC1 (Energy)", marker="o")
sns.lineplot(data=df_mfcc, x=range(len(df_mfcc)), y="MFCC2", label="MFCC2 (Timbre)", marker="s")
sns.lineplot(data=df_mfcc, x=range(len(df_mfcc)), y="MFCC10", label="MFCC10 (Pronunciation)", marker="d")

# Dettagli del grafico
#plt.title("MFCC1, MFCC2, and MFCC10 Across Audio Files", fontsize=16)
plt.xlabel("Audio File Index", fontsize=16)
plt.ylabel("Mel-Frequency Cepstral Coefficients Value", fontsize=16)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend(fontsize=15)
#plt.tight_layout()
plt.savefig(os.path.join(output_folder, "mfcc_combined_plot.png"))
plt.show()
```

```python
import os
import librosa
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# === CONFIGURAZIONE ===
dataset_folder = "/Users/geofargy95/Downloads/BioVid-main/dataset"  # Percorso alla cartella "dataset"
output_folder = "audio_analysis_results"
os.makedirs(output_folder, exist_ok=True)

# === ANALISI AUDIO ===
audio_features = []

# Estensioni supportate per i file video
supported_extensions = (".mp4", ".MP4")  # Aggiungi altre estensioni se necessario

# === ESTRAZIONE DELLE CARATTERISTICHE AUDIO ===
for root, dirs, files in os.walk(dataset_folder):
    for file in files:
        if file.endswith(supported_extensions):  # Controlla se il file ha un'estensione supportata
            file_path = os.path.join(root, file)
            try:
                # Carica il file audio
                y, sr = librosa.load(file_path, sr=None)
                
                # Durata
                duration = librosa.get_duration(y=y, sr=sr)
                
                # Root Mean Square (RMS)
                rms = librosa.feature.rms(y=y).mean()
                
                # Zero Crossing Rate (ZCR)
                zcr = librosa.feature.zero_crossing_rate(y).mean()
                
                # Estrai i primi 13 MFCC
                mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13).mean(axis=1)
                
                # Salva le caratteristiche rilevanti
                audio_features.append({
                    "File": file,
                    "Folder": os.path.basename(root),
                    "Duration": duration,
                    "RMS": rms,
                    "ZCR": zcr,
                    "MFCC1": mfccs[0],
                    "MFCC2": mfccs[1],
                    "MFCC10": mfccs[9],  # Il 10 MFCC
                })
            except Exception as e:
                print(f"Error processing {file_path}: {e}")

print(f"Total number of videos analyzed: {len(audio_features)}")

# Crea un DataFrame con le caratteristiche estratte
df_features = pd.DataFrame(audio_features)
df_features.to_csv(os.path.join(output_folder, "audio_features.csv"), index=False)
print("Audio features saved to audio_features.csv")

# === VISUALIZZAZIONI ===

# 1. Distribuzione delle durate
plt.figure(figsize=(10, 6))
sns.histplot(df_features["Duration"], kde=True, color='blue', bins=25)
plt.title("Distribution of Video Durations", fontsize=25)
plt.xlabel("Duration (seconds)", fontsize=25)
plt.ylabel("Frequency", fontsize=25)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "duration_distribution.png"))
plt.show()

# 2. Boxplot delle durate
plt.figure(figsize=(8, 8))  # Dimensioni della figura
sns.boxplot(df_features["Duration"], color='orange')  # Boxplot con colore specifico

# Titolo (opzionale, decommenta se necessario)
# plt.title("Boxplot of Video Durations", fontsize=25)

# Etichetta dell'asse X
plt.xlabel("Duration (seconds)", fontsize=25)

# Tick dell'asse X
plt.xticks(fontsize=25)

# Tick dell'asse Y
plt.yticks(fontsize=25)

# Layout ottimizzato
#plt.tight_layout()

# Salva il grafico
plt.savefig(os.path.join(output_folder, "duration_boxplot.png"))

# Mostra il grafico
plt.show()

# 2. Boxplot delle durate
#plt.figure(figsize=(8, 8))
#sns.boxplot(x=df_features["Duration"], color='orange')
#plt.xlabel("Duration (seconds)", fontsize=25)
#plt.xticks(fontsize=25)
#plt.yticks(fontsize=25)
#plt.tight_layout()
#plt.savefig(os.path.join(output_folder, "duration_boxplot.png"))
#plt.show()

# 3. Pairplot delle caratteristiche principali
selected_features = ["Duration", "RMS", "ZCR"]
sns.pairplot(df_features[selected_features])
plt.savefig(os.path.join(output_folder, "pairplot_features.png"))
plt.show()

# === REPORT ===
with open(os.path.join(output_folder, "report.txt"), "w") as report:
    report.write("Audio Analysis Report\n")
    report.write("=====================\n")
    report.write(f"Total Files Processed: {len(df_features)}\n")
    report.write("\nSummary Statistics:\n")
    report.write(df_features.describe().to_string())
    report.write("\n\nPlots and features saved in the output folder.\n")

print("Analysis complete. Results saved in:", output_folder)
```
